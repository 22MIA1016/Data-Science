```python
#fast clustering based feature subset selection
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Select relevant features for clustering
features = data.iloc[:, 2:]  # Assuming your data starts from column index 2

# Drop non-numeric columns
features = features.drop(columns=['university'])

# Standardize the features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# Initialize KMeans with desired number of clusters
num_clusters = 3  # You can adjust this number as needed
kmeans = KMeans(n_clusters=num_clusters)

# Fit KMeans to the data
kmeans.fit(scaled_features)

# Get cluster labels for each data point
cluster_labels = kmeans.labels_

# Add cluster labels to the dataframe
data['Cluster'] = cluster_labels

# Calculate centroids of clusters
centroids = scaler.inverse_transform(kmeans.cluster_centers_)  # Scale back centroids

# Plot the features
for i, feature_name in enumerate(features.columns):
    plt.figure(figsize=(10, 6))
    for cluster_num in range(num_clusters):
        plt.subplot(1, num_clusters, cluster_num + 1)
        cluster_data = features[data['Cluster'] == cluster_num][feature_name]
        plt.hist(cluster_data, alpha=0.5, label=f'Cluster {cluster_num}')
        plt.axvline(centroids[cluster_num, i], color='r', linestyle='dashed', linewidth=1, label='Centroid')
        plt.title(feature_name)
        plt.xlabel('Value')
        plt.ylabel('Frequency')
        plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
#relief algorithm
import numpy as np
import pandas as pd

# Drop non-numeric columns
df_numeric = df.drop(['rank', 'university'], axis=1)

# Normalize the data
df_normalized = (df_numeric - df_numeric.mean()) / df_numeric.std()

# Convert dataframe to numpy array
features = df_normalized.values
labels = df['overall_score'].values

def relief_algorithm(features, labels, num_neighbors=5, num_iterations=1000):
    num_instances, num_features = features.shape

    # Initialize weights for features
    weights = np.zeros(num_features)

    for _ in range(num_iterations):
        # Select a random instance
        idx = np.random.randint(num_instances)
        instance = features[idx]
        label = labels[idx]

        # Find the nearest instances of the same and different class
        same_class_indices = np.where(labels == label)[0]
        diff_class_indices = np.where(labels != label)[0]

        nearest_same_class_idx = same_class_indices[np.argsort(np.linalg.norm(features[same_class_indices] - instance, axis=1))[1:num_neighbors+1]]
        nearest_diff_class_idx = diff_class_indices[np.argsort(np.linalg.norm(features[diff_class_indices] - instance, axis=1))[:num_neighbors]]

        # Update feature weights
        if len(nearest_same_class_idx) > 0 and len(nearest_diff_class_idx) > 0:
            weights += np.abs(instance - features[nearest_same_class_idx]).mean(axis=0) - np.abs(instance - features[nearest_diff_class_idx]).mean(axis=0)

    # Normalize weights
    weights /= num_iterations

    return weights

# Call the Relief algorithm function
feature_weights = relief_algorithm(features, labels)

# Print the feature weights along with column names
print("Feature Weights:")
for i, (col_name, weight) in enumerate(zip(df_numeric.columns, feature_weights)):
    print(f"{col_name}: {weight}")
```
