```python
import pandas as pd

df = pd.read_csv('/content/Employee Sample Data.csv')
df
```

```python
import pandas as pd
import matplotlib.pyplot as plt


# Extract the 'Gender' column from the DataFrame
genders = df['Gender']

# Count the occurrences of each gender
gender_counts = genders.value_counts()

# Create bar chart
gender_counts.plot(kind='bar', color='skyblue')

# Add labels and title
plt.xlabel('Gender')
plt.ylabel('Frequency')
plt.title('Distribution of Genders in the Dataset')

# Show plot
plt.show()

```

```python

# Extract the 'Gender' column from the DataFrame
genders = df['Age']

# Count the occurrences of each gender
gender_counts = genders.value_counts()

# Create bar chart
gender_counts.plot(kind='bar', color='skyblue')

# Add labels and title
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Genders in the Dataset')

# Show plot
plt.show()
```

```python
# to fill all the missing values in the dataset using mean of it
numerical_columns = df.select_dtypes(include=['int64', 'float64']).columns
df[numerical_columns] = df[numerical_columns].fillna(df[numerical_columns].mean())


categorical_columns = df.select_dtypes(include=['object']).columns
df[categorical_columns] = df[categorical_columns].fillna(df[categorical_columns].mode().iloc[0])

#  to Display the first few rows of the DataFrame
print(df.head())
```

```python
#to count the number of departments
department_counts = df['Department'].value_counts()
print(department_counts)
```

```python
# calculating the co-relation
df.corr()
```

```python
#Convert 'Exit Date' and 'Hire Date' columns to datetime objects
df['Exit Date'] = pd.to_datetime(df['Exit Date'])
df['Hire Date'] = pd.to_datetime(df['Hire Date'])
df
```
